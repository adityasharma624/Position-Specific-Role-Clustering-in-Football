```python
import numpy as np
import pandas as pd

main_df = pd.read_csv("../dataset/final_with_gmm.csv")
```

```python
top_per_cluster = {}

for k in sorted(main_df["gmm_cluster"].unique()):
    cluster_df = main_df[main_df["gmm_cluster"] == k].head(10)
    top_per_cluster[k] = cluster_df

    print(f"\n=== Cluster {k} — sample 10 players ===")
    print(cluster_df[["Name", "Club", "Division", "Nat", "Age"] if "Age" in cluster_df.columns else ["Name", "Club", "Division"]])
```
