```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../dataset/raw/FM24.csv')
print("Original columns:\n", df.columns.tolist())
```

```python
raw_features = ['Name', 'Club', 'Nat', 'Height', 'Weight', 'Age', 'CA', 'PA', 
'Acc', 'Agi', 'Bal', 'Jum', 'Nat.1', 'Pac', 'Sta', 'Str', 
'Agg', 'Ant', 'Bra', 'Cmp', 'Cnt', 'Dec', 'Det', 'Fla', 
'Ldr', 'OtB', 'Pos', 'Tea', 'Vis', 'Wor', 'Cor', 'Cro', 
'Dri', 'Fin', 'Fir', 'Fre', 'Hea', 'Lon', 
'L Th', 'Mar', 'Pas', 'Pen', 'Tck', 'Tec']
raw_df = df[raw_features].copy()
```

```python
cols = [
    'Acc', 'Agi', 'Bal', 'Jum', 'Nat.1', 'Pac', 'Sta', 'Str',
    'Agg', 'Ant', 'Bra', 'Cmp', 'Cnt', 'Dec', 'Det', 'Fla',
    'Ldr', 'OtB', 'Pos', 'Tea', 'Vis', 'Wor', 'Cor', 'Cro',
    'Dri', 'Fin', 'Fir', 'Fre', 'Hea', 'Lon',
    'L Th', 'Mar', 'Pas', 'Pen', 'Tck', 'Tec'
]

num_df = raw_df[cols].copy()
```

```python
X_norm = (num_df - num_df.mean(axis=1).values.reshape(-1,1)) / num_df.std(axis=1).values.reshape(-1,1)
```

```python
X_norm.to_csv('../../dataset/processed/normalized_features.csv', index=False)
```
