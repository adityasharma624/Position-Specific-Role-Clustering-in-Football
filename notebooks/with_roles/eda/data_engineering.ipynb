```python
import pandas as pd
```

```python
df = pd.read_csv('../../../data/processed/num_data.csv')
df.columns
```

```python
def attribute_scorer(df, core_attributes, opt_attributes):
    core_w=1.0
    opt_w=0.5
    core = [c for c in core_attributes if c in df.columns]
    opt  = [c for c in opt_attributes if c in df.columns]
    num = (df[core].mean(axis=1) * core_w * len(core) +
           df[opt].mean(axis=1)  * opt_w  * len(opt))
    den = core_w * len(core) + opt_w * len(opt)
    return num / den
```

```python
mezzala_a_core = ['Dri', 'Pas', 'Tec', 'Dec', 'OtB', 'Vis', 'Wor', 'Acc']
mezzala_a_opt = ['Fin', 'Fir', 'Ant', 'Com', 'Bal', 'Fla', 'Sta']
df['M_A_Score'] = attribute_scorer(df, mezzala_a_core, mezzala_a_opt)
```
