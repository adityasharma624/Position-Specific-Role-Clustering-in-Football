```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler, Normalizer
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('../../../data/processed/role_scores.csv')
X = df.values
```

```python
preproc = Pipeline([
    ("scaler", StandardScaler()),
    ("row_norm", Normalizer(norm="l2")),
])

X_preprocessed = preproc.fit_transform(X)
```

```python
X_preproc = pd.DataFrame(X_preprocessed, columns=df.columns)
X_preproc.to_csv('../../../data/processed/role_scores_normalized.csv', index=False)
```
