```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler, Normalizer
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('../../../data/processed/role_scores.csv')
```

```python
X_norm = (df - df.mean(axis=1).values.reshape(-1,1)) / df.std(axis=1).values.reshape(-1,1)
```

```python
X_norm.to_csv('../../../data/processed/role_scores_normalized.csv', index=False)
```
