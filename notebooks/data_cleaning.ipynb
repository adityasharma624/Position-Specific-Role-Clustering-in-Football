```python
import pandas as pd
```

```python
df = pd.read_csv('../dataset/raw.csv')
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(' ', '_')
```

```python
print(len(df['Division'].unique()))
```

```python
df_cm = df.loc[df['Position'].str.contains('M (C)', regex=False, na=False)].copy()
df_cm.head()
df_cm.shape
df_cm['Division'] = df_cm['Division'].str.strip()
```

```python
top_5_leagues = [
    "English Premier Division",
    "Spanish First Division",
    "Italian Serie A",
    "Bundesliga",
    "Ligue 1 Uber Eats"
]

df_top5 = df_cm[df_cm['Division'].isin(top_5_leagues)]
df_top5.shape
```

```python
df_top5.columns
df_top5 = df_top5.applymap(
    lambda x: " ".join(x.split()) if isinstance(x, str) else x
)
df_top5.to_csv('../dataset/cleaned_cm_top5.csv', index=False)
```
